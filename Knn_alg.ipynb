```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
```

```python
data = pd.read_csv("wisc_bc_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.drop('id',axis=1,inplace=True)
#for drop row
#data.drop([4],axis=0,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['diagnosis']=data.diagnosis.astype('category')
```

```python
data.describe()
```

```python
data.groupby(['diagnosis']).count()
```

```python
#split data
x=data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
x.head()
```

```python
y.head()
```

```python
#scaled data
#apply zscore for normalizie the data
xscaled=x.apply(zscore)
```

```python
xscaled.describe()
```

```python
x_train,x_test,y_train,y_test=train_test_split(xscaled,y,test_size=0.30,random_state=1)
```

```python
model=KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
#confusion matrix
from sklearn import metrics
print('Confusion matrix')
c=metrics.confusion_matrix(y_test,y_predict)
```

```python
d=pd.DataFrame(c,index=[i for i in["M","B"]],
              columns=[i for i in["Predict M","Predict B"]])
```
